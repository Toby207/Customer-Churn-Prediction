CUSTOMER CHURN PREDICTION

This project aims to develop a classifier for predicting customer churn using a dataset sourced from Kaggle, consisting of 10,000 samples. The primary focus is on the 'exited' variable, indicating whether a customer has churned. Visualization techniques are utilized to understand the data patterns, and machine learning algorithms including Support Vector Machine, Random Forest, Decision Trees, and K-nearest neighbors are employed for prediction. The goal is to provide actionable insights and predictive models to help businesses mitigate customer churn effectively and foster long-term customer satisfaction and organizational growth.

```python
# Importing necessary libraries for label encoding and one-hot encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Instantiating a LabelEncoder object
label = LabelEncoder()

# Encoding the 'Gender' column in the DataFrame 'X' using label encoding
X['Gender'] = label.fit_transform(X['Gender'])

# Printing the first 7 encoded values of the 'Gender' column
print(X['Gender'].head(7))
```

```python
# Encoding the 'Geography' column in the DataFrame 'X' using label encoding
X['Geography'] = label.fit_transform(X['Geography'])

# Printing the first few encoded values of the 'Geography' column
print(X['Geography'].head())

# Displaying the counts of each encoded value in the 'Geography' column
X['Geography'].value_counts()

```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.compose import ColumnTransformer

# Instantiating a ColumnTransformer object to specify which columns to one-hot encode
column_transformer = ColumnTransformer([('encoder', OneHotEncoder(), [1])], remainder='passthrough')

# Performing one-hot encoding on the DataFrame 'X' and converting the result to a NumPy array
X = column_transformer.fit_transform(X)

# Printing the resulting array after one-hot encoding
print(X)
```

```python
# Importing necessary libraries for splitting the data and scaling features
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Splitting the dataset into training and testing sets
# X: features, y: target variable, test_size: proportion of the dataset to include in the test split
# random_state: seed for random number generation to ensure reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
# Instantiating a StandardScaler object to standardize the features
sc_X = StandardScaler()

# Standardizing the training set features
X_train = sc_X.fit_transform(X_train)

# Standardizing the testing set features using the mean and standard deviation from the training set
X_test = sc_X.transform(X_test)
```

```python
#Shape of train and test data
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# Using KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Train Model  
neigh = KNeighborsClassifier(n_neighbors = 4).fit(X_train,y_train)
```

```python
#Prediction
prediction = neigh.predict(X_test)
```

```python
prediction1=pd.DataFrame(prediction)
prediction1.head()
```

```python
# Calculating the accuracy of the model
# Importing the necessary module from scikit-learn for computing accuracy
from sklearn import metrics

# Calculating the accuracy score by comparing the true labels (y_test) with the predicted labels (prediction)
percent1 = metrics.accuracy_score(y_test, prediction)

# Displaying the accuracy score
percent1
```

```python
# Now Using SVM Algorithm 
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=0)
classifier.fit(X_train,y_train)
```

```python
# Making predictions on the testing set using the trained classifier
# The trained classifier (model) predicts the target variable (churn) based on the testing set features (X_test)
y_pred = classifier.predict(X_test)
```
